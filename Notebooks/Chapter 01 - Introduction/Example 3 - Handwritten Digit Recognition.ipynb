# Handwritten Digit Recognition

```python
import matplotlib.pyplot as plt
import numpy as np
from ml_datasets.esl import HandwrittenDigit
from ml_datasets.utils import plot_images
```

```python
digit = HandwrittenDigit(verbose=0)
x_train, y_train, x_test, y_test = digit.load()
```

```python
idx = []
for i in range(10):
    idx_temp = np.where(y_train == i)[0][:6]
    idx.extend(list(idx_temp))
```

```python
x_train_reshaped = x_train[idx, :].reshape(-1, 16, 16)
y_train_reshaped = y_train[idx].reshape(-1, 1)

plt.figure(figsize=(10, 6))
plt.subplots_adjust(hspace=0, wspace=0)
k = 1
for i in range(10):
    k = i + 1
    idxs = np.where(y_train_reshaped == i)[0][:6]
    for j, idx_ in enumerate(idxs):
        ax1 = plt.subplot(6, 10, k, frameon=True)
        ax1.imshow(x_train_reshaped[idx_, ...], cmap='Greys')
        ax1.spines['bottom'].set_color('lime')
        ax1.spines['top'].set_color('lime')
        ax1.spines['left'].set_color('lime')
        ax1.spines['right'].set_color('lime')
        ax1.xaxis.set_visible(False)
        ax1.yaxis.set_visible(False)
        k += 10
```
